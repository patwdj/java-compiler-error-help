```python
import openpyxl

def get_column_values(filename, sheetname, column_letter):
    workbook = openpyxl.load_workbook(filename)
    sheet = workbook[sheetname]

    column_values = []
    for cell in sheet[column_letter]:
        column_values.append(cell.value)

    return column_values

filename = 'dataset.xlsx'
sheetname = 'Java codes + ChatGPT annotation'  # Replace with your sheet name
column_letter = 'C'  # Replace with the desired column letter

snippets = get_column_values(filename, sheetname, column_letter)
```

```python
#remove header and empty code

snippets2 = []

for snippet in snippets[1:]:
    if not snippet == None:
        snippets2.append(snippet) 

snippets = snippets2
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import statistics

def calculate_loc(code_snippets):
    """
    Calculate the lines of code (LOC) for each code snippet,
    excluding empty lines or lines consisting only of whitespace.
    """
    loc_counts = []

    for snippet in code_snippets:
        lines = snippet.strip().split("\n")
        loc_count = sum(1 for line in lines if line.strip() != "")
        loc_counts.append(loc_count)

    return loc_counts

def calculate_mean(loc_counts):
    """
    Calculate the mean (average) of the lines of code (LOC).
    """
    return statistics.mean(loc_counts)

def calculate_median(loc_counts):
    """
    Calculate the median of the lines of code (LOC).
    """
    return statistics.median(loc_counts)

def generate_distribution_visualization(loc_counts):
    """
    Generate a distribution visualization using matplotlib with grey bars
    and display the count of code snippets for each lines of code value.
    """
    # Count the occurrences of each lines of code value
    loc_counter = Counter(loc_counts)

    # Extract the lines of code values and their respective counts
    loc_values = list(loc_counter.keys())
    loc_counts = list(loc_counter.values())

    # Create the bar chart
    plt.bar(loc_values, loc_counts, color='grey')

    plt.xlabel("Lines of Code")
    plt.ylabel("Count")
    plt.title("Distribution of Lines of Code")
    plt.tight_layout()

    # Display the count of code snippets for each lines of code value
    for loc, count in zip(loc_values, loc_counts):
        print(f"{loc} lines of code: {count} code snippet(s)")

    # Save the chart as a PNG image
    plt.savefig("loc_distribution.png", dpi=300)

    # Display the chart
    plt.show()

# Calculate the lines of code (LOC) for each code snippet
loc_counts = calculate_loc(snippets)
mean_loc = calculate_mean(loc_counts)
median_loc = calculate_median(loc_counts)

print("Mean LOC:", mean_loc)
print("Median LOC:", median_loc)

# Generate and display the distribution visualization
generate_distribution_visualization(loc_counts)
```
