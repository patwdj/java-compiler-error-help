```python
rbos = {}
for i in range(len(ids)):
    row = []
    for j in range(len(ids)):
        inside = []
        for k in range(100):
            a = data_Linest[i][k]
            b = data_Linest[j][k]
            rbo_val = rbo.RankingSimilarity(a, b).rbo()
            inside.append(rbo_val)
        row.append(statistics.median(inside))
    rbos[ids[i]] = row
```

```python
import pandas as pd
rbos_pd = pd.DataFrame(rbos, index = ids)
rbos_pd.columns = ids
```

```python
s_data_Linest = data_Linest[0:36]
g_data_Linest = data_Linest[36::]
s_ids = ids[0:36]
g_ids =ids[36::]
```

```python
empty_count_s = {}
empty_count_s_exc_NoSnippet = {}
for i in range(len(s_data_Linest)):
    count = 0
    id_name = s_ids[i]
    item = s_data_Linest[i]
    for j in item:
        if j ==[]:
            count+=1
    empty_count_s[id_name] = count
    if id_name[-9:]!="NoSnippet":
        empty_count_s_exc_NoSnippet[id_name] = count
    
    
empty_count_g = {}
empty_count_g_exc_NoSnippet = {}
for i in range(len(g_data_Linest)):
    count = 0
    id_name = g_ids[i]
    item = g_data_Linest[i]
    for j in item:
        if j ==[]:
            count+=1
    empty_count_g[id_name] = count
    if id_name[-9:]!="NoSnippet":
        empty_count_g_exc_NoSnippet[id_name] = count
```

```python
print(empty_count_s)
print("Average: " , sum(empty_count_s.values())/len(empty_count_s.values()))
print("Average Empty (%): ", sum(empty_count_s.values())/len(empty_count_s.values())/100*100, "%")
```
